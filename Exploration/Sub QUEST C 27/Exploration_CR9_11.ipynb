```python
# ì˜¤ë¥˜ë°œìƒ, ì½”ë©ì—ì„œ í•˜ê¸°(11_2íŒŒì¼ ì°¸ê³ )
import torch 
from diffusers import DiffusionPipeline, DPMSolverMultistepScheduler

device = "cuda"

# íŒŒì´í”„ë¼ì¸ ë¶ˆëŸ¬ì˜¤ê¸°
repo_id = "stabilityai/stable-diffusion-2-base"
pipe = DiffusionPipeline.from_pretrained(repo_id, torch_dtype=torch.float16)

pipe.scheduler = DPMSolverMultistepScheduler.from_config(pipe.scheduler.config)
pipe = pipe.to(device)
```

#### ì—¬ëŸ¬ ê°œì˜ ì´ë¯¸ì§€ ìƒì„±í•˜ê¸°
ì´ë²ˆì—ëŠ” í•˜ë‚˜ì˜ í”„ë¡¬í”„íŠ¸ë¡œ ì—¬ëŸ¬ ê°œì˜ ì´ë¯¸ì§€ë¥¼ ìƒì„±í•´ ë´…ì‹œë‹¤.

ì´ë¯¸ì§€ ì²˜ë¦¬ë¥¼ í•´ì£¼ëŠ” íŒŒì´ì¬ ë¼ì´ë¸ŒëŸ¬ë¦¬ì¸ pillowë¥¼ ì‚¬ìš©í•˜ì—¬ ì—¬ëŸ¬ ê°œì˜ ì´ë¯¸ì§€ë¥¼ ë‹´ì„ í‹€ì„ ë¨¼ì € ë§Œë“¤ì–´ ì¤ë‹ˆë‹¤.

```python
# íŒŒì´ì¬ ì´ë¯¸ì§€ ì²˜ë¦¬ ë¼ì´ë¸ŒëŸ¬ë¦¬ pillow ë¶ˆëŸ¬ì˜¤ê¸° 
from PIL import Image  

# í‹€ ë§Œë“¤ê¸°
def image_grid(imgs, rows, cols):
    assert len(imgs) == rows * cols

    w, h = imgs[0].size  
    grid = Image.new('RGB', size=(cols * w, rows * h))
    grid_w, grid_h = grid.size
    
    for i, img in enumerate(imgs):
        grid.paste(img, box = (i%cols * w, i // cols * h))
    return grid
```
