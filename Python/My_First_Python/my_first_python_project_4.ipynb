### 1단계: 포션 딕셔너리 만들기 📚

당신의 포션 상점이 점점 커지면서 희귀한 포션도 취급하기 시작했습니다. 희귀한 포션을 들여오니 손님들이 포션에 대한 설명이 있었다면 좋겠다고 요구하네요!  
새로운 포션을 리스트에 넣으려고 보니, 포션 이름, 가격, 설명 리스트를 모두 업데이트 해야 합니다.  
여러 리스트로 포션 정보를 관리하는 건 너무 번거롭네요! 이제는 딕셔너리의 시대입니다!  
* 포션 이름을 키로 삼고 [가격, 설명]의 리스트를 값으로 하는 딕셔너리를 만들어보세요.
* 희귀 포션의 정보를 업데이트 해보세요!
    * 예시)
        * 천리안의 성수, 4000 골드, 5분간 적의 약점 노출
        * 크리티컬 블러드, 4500 골드, 5분간 치명타 확률 증가, 체력 5% 감소

```python
# 예시
포션_정보 = {
    '체력 포션': [575, '체력을 회복 시켜주는 물약'],  # [가격, 설명]
    '마나 포션': [700, '마나를 회복 시켜주는 물약'],
    # 더 많은 포션들...
}

```

````python
portion = {
    '체력 포션': [575, '체력을 회복 시켜주는 물약'],  # [가격, 설명]
    '마나 포션': [700, '마나를 회복 시켜주는 물약'],
    '행운 포션': [1000, '아이템 드롭율과 회피율을 높여주는 물약'],
    '투명 포션': [1200, '적에게 내가 보이지 않게 해주는 물약'],
}
````

````python
# 업데이트 하기
#     '천리안의 성수': [4000, '5분간 적의 약점 노출'],
#    '크리티컬 블러드': [4500, '5분간 치명타 확률 증가, 체력 5% 감소']
portion['천리안의 성수'] = [4000, '5분간 적의 약점 노출']
portion['크리티컬 블러드'] = [4500, '5분간 치명타 확률 증가, 체력 5% 감소']
print(portion)
print(portion['체력 포션'][0])
````

### 2단계: 메뉴 출력 함수 만들기

판매하는 목록이 늘어나니, 손님들에게도 판매 목록을 잘 전달할 필요가 생겼습니다! 포션 딕셔너리에 있는 내용을 이쁘게 출력하는 함수를 만들어봅시다!
```python
출력 예시:
======================================판매중인 물약=======================================
물약 이름     : -------------------효과-------------------  가격                  
체력 포션     : 체력을 회복 시켜주는 물약----------------   575 골드
마나 포션     : 마나를 회복 시켜주는 물약----------------   700 골드
행운의 물약    : 30분간 행운이 약간 상승----------------   850 골드
투명 포션     : 10분간 투명 상태를 유지----------------  1990 골드
천리안의 성수   : 5분간 적의 약점 노출------------------  4000 골드
크리티컬 블러드  : 5분간 치명타 확률 증가, 체력 5% 감소-------  4500 골드
==============================================================================================
```

````python
for a in portion.keys():
  print(a)
  print(portion.get(a)[0], portion.get(a)[1])
````

````python
print('='*40+'판매중인 물약'+'='*40+"\n")
print('물약 이름     :','-'*30+'효과'+'-'*30,'가격'+"\n")
for a in portion.keys():
  print(f'{a}     : {portion.get(a)[1]} ------------------------- {portion.get(a)[0]} 골드'+"\n")
print("\n"+'='*44+'캡틴 상점'+'='*44)
# 앞 뒤 분량을 맞추도록 하는 건 GPT에게 물어보자
````

````python
def show():
    print('='*40 + '판매중인 물약' + '='*40 + "\n")
    print('물약 이름     :', '-'*30 + '효과' + '-'*30, '가격' + "\n")

    for name, (price, effect) in portion.items():
        print(f'{name:<12} : {effect:<30} ------------------------- {price} 골드' + "\n")

    print('\n' + '='*44 + '캡틴 상점' + '='*44)
show()
````

### 3단계: 주문 받는 함수 만들기 🛒

이제 중요한 기능들을 함수로 분리해볼까요? 먼저 주문을 받는 함수부터!  
이렇게 하면 나중에 주문 로직을 수정하고 싶을 때 이 함수만 고치면 된답니다! 👌
* 판매중인 포션 정보를 토대로 손님의 주문 목록을 만드는 함수를 구현해봅시다
* 이 함수는 주문 목록을 반환 합니다.
* 함수를 실행해보고, 반환된 주문 목록을 출력해보세요!
```python
# 예시
def 주문_받기(포션_정보):
    주문_목록 = []
    while True:
        # 사용자로부터 포션 이름 입력 받기
        # 0 입력 시 주문 종료
        # 포션 개수 입력 받기
        # 주문 목록에 (포션 이름, 포션 개수) 튜플 추가하기
    return 주문_목록
```



````python
def order():
  order_list = []
  while True:
    a = str(input('구매를 원하는 포션 이름을 띄어쓰기 포함 정확하게 입력해주세요'))
    if not a in portion.keys():
      print(f'고객님이 적어주신 {a}은 저희 캡틴 상점에는 없는 포션입니다. 다시 한번 확인해주세요.')
      continue
    else:
      b = int(input(f'{a}를 몇 개 구매하실 건가요? 숫자만 입력해주세요'))
      order_list.append((a,b))
      print(f'{a} {b}개 장바구니 담기 완료!')
    c = int(input('추가로 구매하실 포션이 있으신가요? 없으면 0, 있으면 아무 숫자를 눌러주세요'))
    if c == 0:
      break
    else:
      continue
  return order_list

my_order=order()
print(my_order)
len(my_order)
````
