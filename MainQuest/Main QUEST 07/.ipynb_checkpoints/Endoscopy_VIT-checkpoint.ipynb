# Transformer 모델 사용해보기

```python
# 필요 라이브러리 로딩
from __future__ import absolute_import, division
from __future__ import print_function, unicode_literals

import os
import time
import shutil
import functools

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib as mpl
mpl.rcParams['axes.grid'] = False
mpl.rcParams['figure.figsize'] = (12,12)

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import matplotlib.image as mpimg
import pandas as pd
from PIL import Image
from IPython.display import clear_output

import tensorflow as tf
import tensorflow_addons as tfa
print(tf.__version__)

from tensorflow.keras import layers
from tensorflow.keras import losses
from tensorflow.keras import models
```

```python
DATASET_PATH = os.path.join(os.getenv('HOME'), 'aiffel', 'medical')

img_dir = os.path.join(DATASET_PATH, "train")
label_dir = os.path.join(DATASET_PATH, "train_labels")

# Split dataset into train data and test data
x_train_filenames = [os.path.join(img_dir, filename) for filename in os.listdir(img_dir)]
x_train_filenames.sort()
y_train_filenames = [os.path.join(label_dir, filename) for filename in os.listdir(label_dir)]
y_train_filenames.sort()

x_train_filenames, x_test_filenames, y_train_filenames, y_test_filenames = \
                    train_test_split(x_train_filenames, y_train_filenames, test_size=0.2)
num_train_examples = len(x_train_filenames)
num_test_examples = len(x_test_filenames)

print("Number of training examples: {}".format(num_train_examples))
print("Number of test examples: {}".format(num_test_examples))
```

```python
# Set hyperparameters
image_size = 256
img_shape = (image_size, image_size, 3)
batch_size = 8
max_epochs = 10
```

```python
def _process_pathnames(fname, label_path):
    img_str = tf.io.read_file(fname)
    img = tf.image.decode_bmp(img_str, channels=3)

    label_img_str = tf.io.read_file(label_path)
    label_img = tf.image.decode_bmp(label_img_str, channels=3)
    label_img = tf.image.rgb_to_grayscale(label_img)

    resize = [image_size, image_size]
    img = tf.image.resize(img, resize)
    label_img = tf.image.resize(label_img, resize)

    scale = 1 / 255.
    img = tf.cast(img, dtype=tf.float32) * scale
    label_img = tf.cast(label_img, dtype=tf.float32) * scale

    return img, label_img

def shift_img(output_img, label_img, width_shift_range, height_shift_range):
    if width_shift_range or height_shift_range:
        if width_shift_range:
                width_shift_range = tf.random.uniform([],
                                                  -width_shift_range * img_shape[1],
                                                  width_shift_range * img_shape[1])
        if height_shift_range:
                height_shift_range = tf.random.uniform([],
                                                   -height_shift_range * img_shape[0],
                                                   height_shift_range * img_shape[0])
        output_img = tfa.image.translate(output_img,
                                         [width_shift_range, height_shift_range])
        label_img = tfa.image.translate(label_img,
                                        [width_shift_range, height_shift_range])
    return output_img, label_img
```

```python
def flip_img(horizontal_flip, tr_img, label_img):
    if horizontal_flip:
        flip_prob = tf.random.uniform([], 0.0, 1.0)
        tr_img, label_img = tf.cond(tf.less(flip_prob, 0.5),
                                lambda: (tf.image.flip_left_right(tr_img), tf.image.flip_left_right(label_img)),
                                lambda: (tr_img, label_img))
    return tr_img, label_img

def _augment(img,
             label_img,
             resize=None,
             scale=1,
             hue_delta=0.1,  # 색상 변화를 위한 기본값 추가
             brightness_delta=0.2,  # 밝기 변화
             contrast_range=(0.8, 1.2),  # 대비 변화 범위
             rotation_range=15,  # 회전 범위 (-15도 ~ 15도)
             horizontal_flip=True,
             width_shift_range=0.05,
             height_shift_range=0.05,
             zoom_range=(0.8, 1.2)):  # 줌 범위 추가
    if resize is not None:
        label_img = tf.image.resize(label_img, resize)
        img = tf.image.resize(img, resize)
  
    if hue_delta:
        img = tf.image.random_hue(img, hue_delta)

    # 밝기와 대비 변화 추가
    img = tf.image.random_brightness(img, brightness_delta)
    img = tf.image.random_contrast(img, contrast_range[0], contrast_range[1])

    # 랜덤 회전
    if rotation_range:
        angle = tf.random.uniform([], -rotation_range, rotation_range)  # 각도를 랜덤으로 선택
        img = tfa.image.rotate(img, angle * tf.constant(np.pi / 180, dtype=tf.float32))
        label_img = tfa.image.rotate(label_img, angle * tf.constant(np.pi / 180, dtype=tf.float32))

    # 좌우 반전
    img, label_img = flip_img(horizontal_flip, img, label_img)

    # 이미지 이동
    img, label_img = shift_img(img, label_img, width_shift_range, height_shift_range)

    # 랜덤 줌
    if zoom_range:
        zoom = tf.random.uniform([], zoom_range[0], zoom_range[1])
        img = tfa.image.transform(img, [zoom, 0, 0, 0, zoom, 0, 0, 0])
        label_img = tfa.image.transform(label_img, [zoom, 0, 0, 0, zoom, 0, 0, 0])

    label_img = tf.cast(label_img, dtype=tf.float32) * scale
    img = tf.cast(img, dtype=tf.float32) * scale
    return img, label_img
```

```python
def get_baseline_dataset(filenames,
                         labels,
                         preproc_fn=functools.partial(_augment),
                         threads=4,
                         batch_size=batch_size,
                         is_train=True):
    num_x = len(filenames)
    dataset = tf.data.Dataset.from_tensor_slices((filenames, labels))
    dataset = dataset.map(_process_pathnames, num_parallel_calls=threads)

    if is_train:
        dataset = dataset.map(preproc_fn, num_parallel_calls=threads)
        dataset = dataset.shuffle(num_x * 10)

    dataset = dataset.batch(batch_size)
    return dataset

train_dataset = get_baseline_dataset(x_train_filenames, y_train_filenames)
train_dataset = train_dataset.repeat()
test_dataset = get_baseline_dataset(x_test_filenames, y_test_filenames, is_train=False)
train_dataset

for images, labels in train_dataset.take(1):
    plt.figure(figsize=(10, 10))
    img = images[0]

    plt.subplot(1, 2, 1)
    plt.imshow(img)

    plt.subplot(1, 2, 2)
    plt.imshow(labels[0, :, :, 0])
    plt.show()

```

```python
class TransformerBlock(layers.Layer):
    def __init__(self, embed_dim, num_heads, mlp_dim, dropout=0.2):  # Dropout 비율 증가
        super().__init__()
        self.att = layers.MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim)
        self.dropout1 = layers.Dropout(dropout)
        self.norm1 = layers.LayerNormalization(epsilon=1e-6)

        self.mlp = tf.keras.Sequential([
            layers.Dense(mlp_dim, activation=tf.nn.gelu),
            layers.Dropout(dropout),  # Dropout 추가
            layers.Dense(embed_dim),
            layers.Dropout(dropout),
        ])
        self.norm2 = layers.LayerNormalization(epsilon=1e-6)

    def call(self, x, training=True):
        attn_output = self.att(x, x)
        x = x + self.dropout1(attn_output, training=training)
        x = self.norm1(x, training=training)

        mlp_output = self.mlp(x, training=training)
        x = x + mlp_output
        x = self.norm2(x, training=training)
        return x
```

```python
class TransformerSegModel(tf.keras.Model):
    def call(self, x):
        # Skip connection 추가: 원본 입력 이미지를 Decoder로 전달
        x_orig = tf.image.resize(x, [self.final_image_size // self.patch_size, self.final_image_size // self.patch_size])
        x = self.patch_embed(x)
        x = x + self.pos_emb

        for blk in self.transformer_blocks:
            x = blk(x)

        h = self.final_image_size // self.patch_size
        w = self.final_image_size // self.patch_size
        x = tf.reshape(x, [-1, h, w, x.shape[-1]])

        # Skip connection으로 원본 feature를 복원 과정에 추가
        x = self.reshape_conv(x + x_orig)  # Skip 연결
        x = self.upsample1(x)
        x = self.upsample2(x)
        x = self.upsample3(x)
        x = self.upsample4(x)
        x = self.out_conv(x)

        return x
```

```python
# Patch Embedding: 크기를 적절히 조정 (예: 8x8)
class PatchEmbedding(layers.Layer):
    def __init__(self, patch_size=8, embed_dim=128):  # Patch 크기: 8x8, Embedding 차원 증가
        super().__init__()
        self.patch_size = patch_size
        self.embed_dim = embed_dim
        self.proj = layers.Dense(embed_dim)

    def call(self, images):
        batch_size = tf.shape(images)[0]
        patches = tf.image.extract_patches(
            images=images,
            sizes=[1, self.patch_size, self.patch_size, 1],
            strides=[1, self.patch_size, self.patch_size, 1],
            rates=[1,1,1,1],
            padding="VALID"
        )
        patch_dims = patches.shape[-1]
        patches = tf.reshape(patches, [batch_size, -1, patch_dims])
        x = self.proj(patches)
        return x
```

```python

# 전체 모델 정의
class TransformerSegModel(tf.keras.Model):
    def __init__(self, image_size=256, patch_size=16, embed_dim=64, num_heads=4, mlp_dim=128, depth=4):
        super().__init__()
        self.patch_size = patch_size
        self.num_patches = (image_size // patch_size) * (image_size // patch_size)

        self.patch_embed = PatchEmbedding(patch_size=patch_size, embed_dim=embed_dim)
        self.pos_emb = self.add_weight("pos_emb", shape=[1, self.num_patches, embed_dim], initializer='zeros')

        self.transformer_blocks = [TransformerBlock(embed_dim, num_heads, mlp_dim) for _ in range(depth)]

        # Decoder: 단순히 feature를 2D map으로 복원하고 upsampling을 통해 segmentation mask 생성
        self.reshape_conv = layers.Conv2DTranspose(64, kernel_size=3, strides=1, padding='same', activation='relu')
        self.upsample1 = layers.UpSampling2D(size=(2,2))
        self.upsample2 = layers.UpSampling2D(size=(2,2))
        self.upsample3 = layers.UpSampling2D(size=(2,2))
        self.upsample4 = layers.UpSampling2D(size=(2,2))
        self.out_conv = layers.Conv2D(1, 1, activation='sigmoid')

        self.final_image_size = image_size

    def call(self, x):
        # x: [batch, H, W, C]
        # Encoder
        x = self.patch_embed(x) # [batch, num_patches, embed_dim]
        x = x + self.pos_emb

        for blk in self.transformer_blocks:
            x = blk(x)

        # x shape: [batch, num_patches, embed_dim]
        # num_patches = (H/patch_size)*(W/patch_size)
        # H/patch_size = W/patch_size = image_size/patch_size
        h = self.final_image_size // self.patch_size
        w = self.final_image_size // self.patch_size
        x = tf.reshape(x, [-1, h, w, x.shape[-1]]) # [batch, H/patch, W/patch, embed_dim]

        # Decoder (간단한 upsampling)
        x = self.reshape_conv(x)
        # 16x16 -> 256x256 복원하려면 16배 확대 필요 = 2x2 upsample을 4번
        x = self.upsample1(x) # 32x32
        x = self.upsample2(x) # 64x64
        x = self.upsample3(x) # 128x128
        x = self.upsample4(x) # 256x256
        x = self.out_conv(x)  # [batch, 256, 256, 1]

        return x
```

```python

def dice_coeff(y_true, y_pred):
    smooth = 1e-10
    # Flatten
    y_true_f = tf.reshape(y_true, [-1])
    y_pred_f = tf.reshape(y_pred, [-1])
    intersection = tf.reduce_sum(y_true_f * y_pred_f)
    score = (2. * intersection + smooth) / (tf.reduce_sum(tf.square(y_true_f)) + \
                                            tf.reduce_sum(tf.square(y_pred_f)) + smooth)
    return score

def dice_loss(y_true, y_pred):
    loss = 1 - dice_coeff(y_true, y_pred)
    return loss

def bce_dice_loss(y_true, y_pred):
    loss = tf.reduce_mean(losses.binary_crossentropy(y_true, y_pred)) + dice_loss(y_true, y_pred)
    return loss
```

```python

transformer_model = TransformerSegModel(image_size=image_size, patch_size=16, embed_dim=64, num_heads=4, mlp_dim=128, depth=4)
optimizer = tf.keras.optimizers.Adam()

transformer_model.compile(optimizer=optimizer, loss=bce_dice_loss, metrics=[dice_loss])

predictions_test = transformer_model(tf.random.normal([batch_size, image_size, image_size, 3]))
print(predictions_test.shape)
transformer_model.summary()

checkpoint_dir_transformer = os.path.join(DATASET_PATH,'train_ckpt/transformer_model')
if not os.path.isdir(checkpoint_dir_transformer):
    os.makedirs(checkpoint_dir_transformer)
```

```python

cp_callback_transformer = tf.keras.callbacks.ModelCheckpoint(checkpoint_dir_transformer,
                                                 save_weights_only=True,
                                                 monitor='val_loss',
                                                 mode='auto',
                                                 save_best_only=True,
                                                 verbose=0)

cos_decay = tf.keras.experimental.CosineDecay(1e-3, max_epochs)
lr_callback = tf.keras.callbacks.LearningRateScheduler(cos_decay, verbose=1)
```

```python

# 모델 학습
history_transformer = transformer_model.fit(train_dataset,
                    epochs=max_epochs,
                    steps_per_epoch=num_train_examples//batch_size,
                    validation_data=test_dataset,
                    validation_steps=num_test_examples//batch_size,
                    callbacks=[cp_callback_transformer, lr_callback]
                    )

```

```python
def mean_iou(y_true, y_pred, num_classes=2):
    y_true_f = tf.reshape(y_true, [-1])
    y_pred_f = tf.reshape(y_pred, [-1])

    y_true_f = tf.cast(tf.round(y_true_f), dtype=tf.int32).numpy()
    y_pred_f = tf.cast(tf.round(y_pred_f), dtype=tf.int32).numpy()

    labels = list(range(num_classes))
    current = confusion_matrix(y_true_f, y_pred_f, labels=labels)

    intersection = np.diag(current)
    ground_truth_set = current.sum(axis=1)
    predicted_set = current.sum(axis=0)
    union = ground_truth_set + predicted_set - intersection

    IoU = intersection / union.astype(np.float32)
    return np.mean(IoU)
```

```python

# 모델 성능 테스트
mean_scores = []
for images, labels in test_dataset:
    preds = transformer_model(images, training=False)
    m = mean_iou(labels, preds)
    mean_scores.append(m)

mean_score = np.mean(mean_scores)
print("mean_iou: {}".format(mean_score))

# batch_size 4 기준 mean_iou: 0.5635994712160207
# batch_size 8 기준 mean_iou: 0.5471973583522203
```

```python
def evaluate_mean_iou(dataset, model, num_classes=2):
    """
    주어진 데이터셋과 모델을 사용하여 mean_iou를 계산합니다.
    
    Args:
        dataset: 평가할 데이터셋 (test_dataset 등)
        model: 학습된 TensorFlow 모델
        num_classes: 클래스의 개수 (기본값: 2)
        
    Returns:
        mean_iou: 데이터셋에 대한 평균 IoU 점수
    """
    mean_scores = []
    for images, labels in dataset:
        # 예측 생성
        preds = model(images, training=False)
        # mean_iou 계산
        m = mean_iou(labels, preds, num_classes)
        mean_scores.append(m)
    
    # 평균 IoU 계산
    return np.mean(mean_scores)

# 고정 학습률 모델에 대한 mean_iou 계산
mean_iou_fixed_lr = evaluate_mean_iou(test_dataset, transformer_model)
print(f"Mean IoU after fixed learning rate training: {mean_iou_fixed_lr}")

# 성능 더 떨어짐
```
